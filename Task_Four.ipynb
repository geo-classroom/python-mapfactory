```python
#classify choropleth map into 3 classes
#Define map properties
column='POP_EST'
k_classes=3
scheme="natural_breaks"
cmap = 'Dark2'
figsize=(20,20)

#Create function for choropleth map with 7 classes
def createMap( df,column,scheme,cmap,figsize):
    map = df.plot( column=column, scheme=scheme,legend=True,cmap = 'Dark2',k=k_classes,figsize=(20,20))
    return map
#Call the function
createMap(df,column,scheme,cmap,figsize)
```
