```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import mapclassify
```

```python
## Function to create proportional map
def proportion_map(base_map_path, point_path, column_name, map_title, circle_colour):
  # Add a polygon of the area from which the points will be referenced 
  #base_map_path = 'path to polygon layer'
  base_map_df = gpd.read_file(base_map_path)

  # path to the points you want to display/measure (make sure its CRS is identical to the polygon layer)
  #point_path = 'path to point layer'
  df = gpd.read_file(point_path)
  df
  basemap = base_map_df.plot(color = 'None', edgecolor = 'black')

  #marker - math (based on the value you want to display) should be used to define how the size of the point circle should be defined 
  proport_map = df.plot(ax=basemap, marker='.', edgecolors='b',figsize=(50, 40), column=column_name, color = 'None', edgecolor = circle_colour, k=2, markersize=df['POP_MAX']/df['MAX_AREAKM'])
  proport_map.set_title(map_title)
  return proport_map
```
