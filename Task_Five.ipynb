```python
#function to create choropleth map
#Define map properties
column='POP_EST'
k_classes=5
scheme="quantiles"
cmap = 'Dark2'
figsize=(20,20)

#Create function for choropleth map with 7 classes
def choropleth( df,column,scheme,cmap,figsize):
    map = df.plot( column=column, scheme=scheme,legend=True,cmap = 'Dark2',k=k_classes,figsize=(20,20))
    return map
#Call the function
choropleth(df,column,scheme,cmap,figsize)
```
