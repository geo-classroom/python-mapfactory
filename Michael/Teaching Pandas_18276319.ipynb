```python
#Task 4
#create a function and add a column of the book prices in Rands.

def rands(x):
    x = x * 18.37
    return x

books['rands(R)']= books['price'].apply(lambda x: rands(x))

books.head(3)
```

```python
#Task 3
#The most expensive book on Amazon? (note: price is already in dollars)

df = books.loc[books['price'].idxmax()]
df
```

```python
#Task 5
#create a function that classifies the length of the book by its pages. Use a 200 page interval and make 3 classes (short, average, long).
#thereafter show the title, number of pages and the length read rating in a dataframe.

def classify(x):
    if x >0 and x <= 200:
        return "short read"
    elif x > 200 and x <= 400:
        return "average read"
    elif x > 400:
        return "long read"
        
books['read_rating'] = books['pages'].apply(lambda x: classify(x))

books[['title','pages','read_rating']]
```
